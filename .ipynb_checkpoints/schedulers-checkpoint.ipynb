```python
def FCFS(processes):
    processes.sort(key=lambda x: x[1])
    fcfs = deque()
    
    clock = 0
    time_left = [list(process) for process in processes]
    order = []
    time = []
    turnaround = []
    waiting = []
    response = [] #todo

    max_cpu_time = processes[0][1] + sum([process[2] for process in processes])
    arrived = 0
    
    while(clock != max_cpu_time):
        if(arrived != len(processes) and clock == time_left[arrived][1]):
            fcfs.append(time_left[arrived])
            arrived = arrived + 1
        clock = clock+1
        if(len(fcfs) != 0):
            curr_process = fcfs.popleft()
            curr_process[2] = curr_process[2]-1
            if(curr_process[2] == 0):
                order.append(curr_process[0])
                time.append(clock)
                turnaround.append([(curr_process[0],clock-curr_process[1])])
                waiting.append([(curr_process[0],clock-curr_process[1]-curr_process[2])])
            else:
                fcfs.appendleft(curr_process)
    print('FCFS Order:',order)
    print('FCFS Time:',time)
    return order, time
```

```python
def SJF(processes,preemption=False):
    processes.sort(key=lambda x: x[1])
    sjf = deque()
    time_left = [list(process) for process in processes]
    order = [] #process order
    time = [] #process time for each process in order
    turnaround = []
    waiting = []
    response = [] #todo
    
    clock = 0
    max_cpu_time = processes[0][1] + sum([process[2] for process in processes])
    arrived = 0
    while(clock != max_cpu_time):
        if(arrived != len(processes) and clock == time_left[arrived][1]):
            if(preemption == False):
                sjf.append(time_left[arrived])
                if(len(sjf) > 1):
                    curr_process = sjf.popleft()
                    p_queuing = list(sjf)
                    p_queuing.sort(key=lambda x: x[2])
                    while(len(sjf) != 0):
                        sjf.pop()
                    for p in p_queuing:
                        sjf.append(p)
                    sjf.appendleft(curr_process)
            elif(preemption == True):
                sjf.append(time_left[arrived])
                if(len(sjf) > 1):
                    curr_process = sjf.popleft()
                    curr_process_id = curr_process[0]
                    sjf.appendleft(curr_process)
                    p_queuing = list(sjf)
                    p_queuing.sort(key=lambda x: x[2])
                    while(len(sjf) != 0):
                        sjf.pop()
                    for p in p_queuing:
                        sjf.append(p)
                    new_process = sjf.popleft()
                    new_process_id = new_process[0]
                    sjf.appendleft(new_process)
                    if(curr_process_id != new_process_id):
                        order.append(curr_process_id)
                        time.append(clock)
            arrived = arrived + 1
        clock = clock+1
        if(len(sjf) != 0):
            curr_process = sjf.popleft()
            curr_process[2] = curr_process[2]-1
            if(curr_process[2] == 0):
                order.append(curr_process[0])
                time.append(clock)
                turnaround.append([(curr_process[0],clock-curr_process[1])])
                waiting.append([(curr_process[0],clock-curr_process[1]-curr_process[2])])
            else:
                sjf.appendleft(curr_process)
    if preemption:
        print('SJF(Preemptive) Order:',order)
        print('SJF(Preemptive) Time:',time)
        print('SJF(Preemptive) Turnaround:',turnaround)
        print('SJF(Preemptive) Waiting Time:',waiting)
    else:
        print('SJF(Non-Preemptive) Order:',order)
        print('SJF(Non-Preemptive) Time:',time)
        print('SJF(Non-Preemptive) Turnaround:',turnaround)
        print('SJF(Non-Preemptive) Waiting Time:',waiting)
    return order, time
```

```python
def RR(processes, qTime):
    processes.sort(key=lambda x: x[1])
    rr = deque()
    
    time_left = [list(process) for process in processes]
    order = [] #process order
    time = [] #process time for each process in order
    turnaround = []
    waiting = []
    response = [] #todo
    
    clock = 0
    max_cpu_time = processes[0][1] + sum([process[2] for process in processes])
    arrived = 0
    
    def find_cpu_burst(process_no):
        for process in processes:
            if(process[0] == process_no):
                return process[2]
        return None
    
    while(clock != max_cpu_time):
        if(arrived != len(processes) and clock == time_left[arrived][1]):
            rr.append(time_left[arrived])
            arrived = arrived + 1
        clock=clock+1
        if(len(rr) != 0):
            curr_process = rr.popleft()
            curr_process[2] = curr_process[2]-1
            process_burst = find_cpu_burst(curr_process[0])
            remainder = (process_burst - curr_process[2]) % qTime
            if(curr_process[2] == 0):
                order.append(curr_process[0])
                time.append(clock)
                turnaround.append([(curr_process[0],clock-curr_process[1])])
                waiting.append([(curr_process[0],clock-curr_process[1]-curr_process[2])])
            elif(remainder == 0):
                rr.append(curr_process)
                order.append(curr_process[0])
                time.append(clock)
                turnaround.append([(curr_process[0],clock-curr_process[1])])
                waiting.append([(curr_process[0],clock-curr_process[1]-curr_process[2])])
            else:
                rr.appendleft(curr_process)
    print('RR Order:',order)
    print('RR Time:',time)
    print('RR Turnaround:',turnaround)
    print('RR Waiting Time:',waiting)
    return order, time
```

```python
def PQ(processes, preemption=False):
    processes.sort(key=lambda x: x[1])
    pq = deque()
    time_left = [list(process) for process in processes]
    order = [] #process order
    time = [] #process time for each process in order
    turnaround = []
    waiting = []
    response = [] #todo
    
    clock = 0
    max_cpu_time = processes[0][1] + sum([process[2] for process in processes])
    arrived = 0
    while(clock != max_cpu_time):
        if(arrived != len(processes) and clock == time_left[arrived][1]):
            if(preemption == False):
                pq.append(time_left[arrived])
                if(len(pq) > 1):
                    curr_process = pq.popleft()
                    p_queuing = list(pq)
                    p_queuing.sort(key=lambda x: x[3])
                    while(len(pq) != 0):
                        pq.pop()
                    for p in p_queuing:
                        pq.append(p)
                    pq.appendleft(curr_process)
            elif(preemption == True):
                pq.append(time_left[arrived])
                if(len(pq) > 1):
                    curr_process = pq.popleft()
                    curr_process_id = curr_process[0]
                    pq.appendleft(curr_process)
                    p_queuing = list(pq)
                    p_queuing.sort(key=lambda x: x[3])
                    while(len(pq) != 0):
                        pq.pop()
                    for p in p_queuing:
                        pq.append(p)
                    new_process = pq.popleft()
                    new_process_id = new_process[0]
                    pq.appendleft(new_process)
                    if(curr_process_id != new_process_id):
                        order.append(curr_process_id)
                        time.append(clock)
            arrived = arrived + 1
        clock = clock+1
        if(len(pq) != 0):
            curr_process = pq.popleft()
            curr_process[2] = curr_process[2]-1
            if(curr_process[2] == 0):
                order.append(curr_process[0])
                time.append(clock)
                turnaround.append([(curr_process[0],clock-curr_process[1])])
                waiting.append([(curr_process[0],clock-curr_process[1]-curr_process[2])])
            else:
                pq.appendleft(curr_process)
    if preemption:
        print('PQ(Preemptive) Order:',order)
        print('PQ(Preemptive) Time:',time)
        print('PQ(Preemptive) Turnaround:',turnaround)
        print('PQ(Preemptive) Waiting Time:',waiting)
    else:
        print('PQ(Non-Preemptive) Order:',order)
        print('PQ(Non-Preemptive) Time:',time)
        print('PQ(Non-Preemptive) Turnaround:',turnaround)
        print('PQ(Non-Preemptive) Waiting Time:',waiting)
    return order,time
```

```python
def MLFQ(processes,qFirst,qSecond,preemption=False):
    processes.sort(key=lambda x: x[1])
    q1 = deque()
    q2 = deque()
    fcfs = deque()
    
    time_left = [list(process) for process in processes]
    order = [] #process order
    time = [] #process time for each process in order
    turnaround = []
    waiting = []
    response = [] #todo
    clock = 0
    max_cpu_time = processes[0][1] + sum([process[2] for process in processes])
    arrived = 0
    
    def find_cpu_burst(process_no):
        for process in processes:
            if(process[0] == process_no):
                return process[2]
        return None
    
    while(clock != max_cpu_time):
        if preemption:
            if(arrived != len(processes) and clock == time_left[arrived][1]):
                q1.append(time_left[arrived])
                arrived = arrived + 1
            clock=clock+1
            if(len(q1) != 0):
                curr_process = q1.popleft()
                curr_process[2] = curr_process[2]-1
                process_burst = find_cpu_burst(curr_process[0])
                if(curr_process[2] == 0):
                    order.append(curr_process[0])
                    time.append(clock)
                    turnaround.append([(curr_process[0],clock-curr_process[1])])
                    waiting.append([(curr_process[0],clock-curr_process[1]-curr_process[2])])
                elif(process_burst - curr_process[2] == qFirst):
                    q2.append(curr_process)
                    order.append(curr_process[0])
                    time.append(clock)
                else:
                    q1.appendleft(curr_process)

            elif(len(q1) == 0 and len(q2) != 0):
                curr_process = q2.popleft()
                curr_process[2] = curr_process[2]-1
                process_burst = find_cpu_burst(curr_process[0])
                if(curr_process[2] == 0):
                    order.append(curr_process[0])
                    time.append(clock)
                    turnaround.append([(curr_process[0],clock-curr_process[1])])
                    waiting.append([(curr_process[0],clock-curr_process[1]-curr_process[2])])
                elif(process_burst - curr_process[2] == qFirst + qSecond):
                    fcfs.append(curr_process)
                    order.append(curr_process[0])
                    time.append(clock)
                else:
                    q2.appendleft(curr_process)

            elif(len(q1) == 0 and len(q2) == 0 and len(fcfs) != 0):
                curr_process = fcfs.popleft()
                curr_process[2] = curr_process[2]-1
                if(curr_process[2] == 0):
                    order.append(curr_process[0])
                    time.append(clock)
                    turnaround.append([(curr_process[0],clock-curr_process[1])])
                    waiting.append([(curr_process[0],clock-curr_process[1]-curr_process[2])])
                else:
                    fcfs.appendleft(curr_process)
        elif(preemption==False):
            while(arrived != len(processes) and clock >= time_left[arrived][1]):
                q1.append(time_left[arrived])
                arrived = arrived + 1
            if(len(q1) != 0):
                curr_process = q1.popleft()
                if(curr_process[2]<qFirst):
                    diff = curr_process[2]
                else:
                    diff = qFirst
                curr_process[2] = curr_process[2]-diff
                process_burst = find_cpu_burst(curr_process[0])
                clock = clock + diff
                if(curr_process[2] == 0):
                    order.append(curr_process[0])
                    time.append(clock)
                    turnaround.append([(curr_process[0],clock-curr_process[1])])
                    waiting.append([(curr_process[0],clock-curr_process[1]-curr_process[2])])
                else:
                    q2.append(curr_process)
                    order.append(curr_process[0])
                    time.append(clock)
            elif(len(q1) == 0 and len(q2) != 0):
                curr_process = q2.popleft()
                if(curr_process[2]<qSecond):
                    diff = curr_process[2]
                else:
                    diff = qSecond
                curr_process[2] = curr_process[2]-diff
                process_burst = find_cpu_burst(curr_process[0])
                clock = clock + diff
                if(curr_process[2] == 0):
                    order.append(curr_process[0])
                    time.append(clock)
                    turnaround.append([(curr_process[0],clock-curr_process[1])])
                    waiting.append([(curr_process[0],clock-curr_process[1]-curr_process[2])])
                else:
                    fcfs.append(curr_process)
                    order.append(curr_process[0])
                    time.append(clock)
            elif(len(q1) == 0 and len(q2) == 0 and len(fcfs) != 0):
                curr_process = fcfs.popleft()
                diff = curr_process[2]
                clock = clock + diff
                curr_process[2] = curr_process[2]- diff
                if(curr_process[2] == 0):
                    order.append(curr_process[0])
                    time.append(clock)
                    turnaround.append([(curr_process[0],clock-curr_process[1])])
                    waiting.append([(curr_process[0],clock-curr_process[1]-curr_process[2])])
            else:
                clock=clock+1
    turnaround.sort(key=lambda x:x[0])
    waiting.sort(key=lambda x:x[0])
    #response.sort(key=lambda x:x[0])
    if preemption:
        print('MLFQ(Preemptive) Order:',order)
        print('MLFQ(Preemptive) Time:',time)
        print('MLFQ(Preemptive) Turnaround:',turnaround)
        print('MLFQ(Preemptive) Waiting Time:',waiting)
        #print('MLFQ(Preemptive) Response Time:',response)
    else:
        print('MLFQ(Non-Preemptive) Order:',order)
        print('MLFQ(Non-Preemptive) Time:',time)
        print('MLFQ(Non-Preemptive) Turnaround:',turnaround)
        print('MLFQ(Non-Preemptive) Waiting Time:',waiting)
        #print('MLFQ(Non-Preemptive) Response Time:',response)
    return order,time
```
